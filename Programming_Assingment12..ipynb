```python
1. Write a Python program to Extract Unique values dictionary values?

Ans-

Sample dictionary
sample_dict = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 2,
    'e': 1
}

Extract unique values from dictionary
unique_values = list(set(sample_dict.values()))

print("Unique values from dictionary:", unique_values)

2. Write a Python program to find the sum of all items in a dictionary?

Ans-

Sample dictionary
sample_dict = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40,
    'e': 50
}

Calculate the sum of all items in the dictionary
total_sum = sum(sample_dict.values())

print("Sum of all items in the dictionary:", total_sum)


3. Write a Python program to Merging two Dictionaries?

Ans-

Sample dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

Merge the dictionaries using update() method
merged_dict = dict1.copy()
merged_dict.update(dict2)

print("Merged dictionary (using update()):", merged_dict)


4. Write a Python program to convert key-values list to flat dictionary?

Ans-

Sample list of key-value pairs
key_value_list = [('a', 1), ('b', 2), ('c', 3), ('d', 4)]

Convert key-values list to flat dictionary
flat_dict = dict(key_value_list)

print("Flat dictionary:", flat_dict)



5. Write a Python program to insertion at the beginning in OrderedDict?

Ans-

from collections import OrderedDict

Sample ordered dictionary
ordered_dict = OrderedDict([('b', 2), ('c', 3), ('d', 4)])

Key-value pair to be inserted at the beginning
new_key = 'a'
new_value = 1

Create a new OrderedDict with the new key-value pair
new_dict = OrderedDict([(new_key, new_value)])

Insert the new key-value pair at the beginning of the existing OrderedDict
new_dict.update(ordered_dict)

print("OrderedDict after insertion at the beginning:")
print(new_dict)




6. Write a Python program to check order of character in string using OrderedDict()?

Ans-

from collections import OrderedDict

def check_order(input_string, pattern):
    Create an OrderedDict to store the order of characters in the input string
    char_order = OrderedDict.fromkeys(input_string)

    Iterate through the pattern and check if characters are in the same order in the OrderedDict
    pattern_index = 0
    for char in char_order:
        if pattern_index < len(pattern) and char == pattern[pattern_index]:
            pattern_index += 1

    If all characters in the pattern are found in the same order, return True, else False
    return pattern_index == len(pattern)

Example usage
input_string = "hello world"
pattern = "hlo"
result = check_order(input_string, pattern)

if result:
    print("Characters in the pattern are in the same order in the input string.")
else:
    print("Characters in the pattern are NOT in the same order in the input string.")


7. Write a Python program to sort Python Dictionaries by Key or Value?

Ans
```
