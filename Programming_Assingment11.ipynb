```python
1. Write a Python program to find words which are greater than given length k?

Ans
    def find_words_greater_than_length(sentence, k):
        words = sentence.split()
        result = []
        for word in words:
            if len(word) > k:
                result.append(word)
        return result

    # Example usage
    sentence = input("Enter a sentence: ")
    k = int(input("Enter the minimum word length: "))
    result_words = find_words_greater_than_length(sentence, k)
    print(f"Words greater than length {k} are: {result_words}")


2. Write a Python program for removing i-th character from a string?

Ans
    def remove_ith_character(input_string, i):
        # Check if the index is valid
        if i < 0 or i >= len(input_string):
            return "Invalid index"
        else:
            # Slice the string to remove the i-th character
            result_string = input_string[:i] + input_string[i + 1:]
            return result_string

    # Example usage
    input_string = input("Enter a string: ")
    index = int(input("Enter the index of the character to remove: "))

    modified_string = remove_ith_character(input_string, index)
    print("String after removing the i-th character:", modified_string)


3. Write a Python program to split and join a string?

Ans
    def split_string(input_string):
        # Split the input string into a list of words
        words = input_string.split()
        return words

        # Example usage
        input_string = input("Enter a string: ")
        split_words = split_string(input_string)
        print("List of words:", split_words)
        
        
    def join_words(word_list):
        # Join the list of words into a single string
        joined_string = ' '.join(word_list)
        return joined_string

        # Example usage
        joined_string = join_words(split_words)
        print("String after joining:", joined_string)



4. Write a Python to check if a given string is binary string or not?

Ans
    def is_binary_string(input_string):
    # Set of valid binary digits
    binary_digits = {'0', '1'}
    
    # Check if all characters in the input string are binary digits
    for char in input_string:
        if char not in binary_digits:
            return False
    return True

    # Example usage
    input_string = input("Enter a string: ")

    if is_binary_string(input_string):
        print("The given string is a binary string.")
    else:
        print("The given string is not a binary string.")


5. Write a Python program to find uncommon words from two Strings?

Ans
    def find_uncommon_words(str1, str2):
    # Split the strings into words
    words_str1 = set(str1.split())
    words_str2 = set(str2.split())
    
    # Find uncommon words in each string
    uncommon_words_str1 = words_str1 - words_str2
    uncommon_words_str2 = words_str2 - words_str1
    
    # Combine the uncommon words from both strings
    uncommon_words = list(uncommon_words_str1.union(uncommon_words_str2))
    return uncommon_words

    # Example usage
    string1 = input("Enter the first string: ")
    string2 = input("Enter the second string: ")

    uncommon_words = find_uncommon_words(string1, string2)
    print("Uncommon words:", uncommon_words)


6. Write a Python to find all duplicate characters in string?

Ans
    def find_duplicate_characters(input_string):
    # Create an empty dictionary to store characters and their counts
    char_count = {}
    duplicates = []

    # Count the occurrences of each character in the input string
    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1

    # Find characters with count more than 1 (duplicates)
    for char, count in char_count.items():
        if count > 1:
            duplicates.append(char)

    return duplicates

    # Example usage
    input_string = input("Enter a string: ")
    duplicate_chars = find_duplicate_characters(input_string)

    if duplicate_chars:
        print("Duplicate characters in the string:", duplicate_chars)
    else:
        print("No duplicate characters found in the string.")


7. Write a Python Program to check if a string contains any special character?

Ans
    import re

    def contains_special_characters(input_string):
        # Regular expression to match special characters
        special_characters_regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
    
        # Check if the input string contains any special characters
        if special_characters_regex.search(input_string) is not None:
            return True
        else:
            return False

# Example usage
input_string = input("Enter a string: ")

if contains_special_characters(input_string):
    print("The string contains special characters.")
else:
    print("The string does not contain any special characters.")
```
