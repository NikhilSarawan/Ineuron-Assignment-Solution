```python
1. Write a Python program to convert kilometers to miles?

Ans
    Function to convert kilometers to miles
    def kilometers_to_miles(kilometers):
    miles = kilometers * 0.621371
    return miles

    Taking input for kilometers from the user
    kilometers = float(input("Enter distance in kilometers: "))

    Converting kilometers to miles
    miles = kilometers_to_miles(kilometers)

    Printing the result
    print(kilometers, "kilometers is equal to", miles, "miles")

2. Write a Python program to convert Celsius to Fahrenheit?

Ans
    Function to convert Celsius to Fahrenheit
    def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

    Taking input for temperature in Celsius from the user
    celsius = float(input("Enter temperature in Celsius: "))

    Converting Celsius to Fahrenheit
    fahrenheit = celsius_to_fahrenheit(celsius)

    Printing the result
    print(celsius, "Celsius is equal to", fahrenheit, "Fahrenheit")

    

3. Write a Python program to display calendar?

Ans
    import calendar

    Taking input for month and year from the user
    year = int(input("Enter the year (e.g., 2023): "))
    month = int(input("Enter the month (1-12): "))

    Displaying the calendar for the specified month and year
    cal = calendar.month(year, month)
    print("Calendar for", calendar.month_name[month], year, ":\n")
    print(cal)

    

4. Write a Python program to solve quadratic equation?

Ans
    import math

    Function to solve the quadratic equation
    def solve_quadratic(a, b, c):
    discriminant = b**2 - 4*a*c
    
    Checking if the discriminant is non-negative
    if discriminant >= 0:
        Calculating the solutions
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    else:
    If the discriminant is negative, no real solutions exist
        return "No real solutions"

    Taking input for coefficients from the user
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
    c = float(input("Enter the coefficient c: "))

    Solving the quadratic equation
    roots = solve_quadratic(a, b, c)

    Printing the solutions
    print("Solutions:", roots)


5. Write a Python program to swap two variables without temp variable?

Ans
    Taking input for the first variable
    variable1 = input("Enter the first variable: ")

    Taking input for the second variable
    variable2 = input("Enter the second variable: ")

    Printing the values before swapping
    print("Before swapping: ")
    print("Variable 1 =", variable1)
    print("Variable 2 =", variable2)

    Swapping the variables without a temporary variable using arithmetic operations
    variable1 = int(variable1) + int(variable2)
    variable2 = int(variable1) - int(variable2)
    variable1 = int(variable1) - int(variable2)

    Printing the values after swapping
    print("\nAfter swapping: ")
    print("Variable 1 =", variable1)
    print("Variable 2 =", variable2)

    
```
