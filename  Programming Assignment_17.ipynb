```python
Question1. Create a function that takes three arguments a, b, c and returns the sum of the
numbers that are evenly divided by c from the range a, b inclusive.


Ans-

def sum_divisible_by_c(a, b, c):
    total_sum = 0
    for number in range(a, b + 1):
        if number % c == 0:
            total_sum += number
    return total_sum

Example usage
a = int(input("Enter the lower bound of the range: "))
b = int(input("Enter the upper bound of the range: "))
c = int(input("Enter the divisor: "))

result = sum_divisible_by_c(a, b, c)
print(f"The sum of numbers between {a} and {b} that are divisible by {c} is: {result}")


Question2. Create a function that returns True if a given inequality expression is correct and
False otherwise.

Ans-

def is_inequality_correct(expression):
    try:
        Use eval() to evaluate the expression
        return eval(expression)
    except:
        Handle invalid expressions
        return False

Example usage
inequality_expression = input("Enter an inequality expression: ")

result = is_inequality_correct(inequality_expression)
if result:
    print(f"The inequality expression '{inequality_expression}' is correct.")
else:
    print(f"The inequality expression '{inequality_expression}' is incorrect.")



Question3. Create a function that replaces all the vowels in a string with a specified character.

Ans-

def replace_vowels(input_string, replacement_char):
    vowels = "aeiouAEIOU"
    Use a list comprehension to replace vowels with the specified character
    modified_string = ''.join(replacement_char if char in vowels else char for char in input_string)
    return modified_string

Example usage
input_string = input("Enter a string: ")
replacement_char = input("Enter the character to replace vowels with: ")

modified_string = replace_vowels(input_string, replacement_char)
print(f"Modified string: {modified_string}")



Question4. Write a function that calculates the factorial of a number recursively.

Ans-

def factorial_recursive(n):
    Base case: factorial of 0 is 1
    if n == 0:
        return 1
    Recursive case: n! = n * (n-1)!
    else:
        return n * factorial_recursive(n - 1)

Example usage
number = int(input("Enter a number to calculate its factorial: "))

Call the recursive function
result = factorial_recursive(number)
print(f"The factorial of {number} is {result}")



Question 5
Hamming distance is the number of characters that differ between two strings.

Ans-

def hamming_distance(str1, str2):
    Check if the strings are of equal length
    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length")
    
    Compute Hamming distance
    distance = 0
    for i in range(len(str1)):
        if str1[i] != str2[i]:
            distance += 1
    return distance

Example usage
string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")

try:
    distance = hamming_distance(string1, string2)
    print(f"Hamming Distance: {distance}")
except ValueError as e:
    print(e)
```
