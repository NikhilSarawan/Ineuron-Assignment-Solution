```python
##1. Assign the value 7 to the variable guess_me. Then, write the conditional tests (if, else, and elif) 
    #to print the string 'too low' if guess_me is less than 7, 'too high' if greater than 7, and 'just right' if equal to 7.

Ans-.Here is how you can assign the value 7 to the variable guess_me and write the conditional tests to print the appropriate,
    string based on the value of guess_me:
    
    # Assign the value 7 to guess_me
    guess_me = 7

    # Write conditional tests to print the appropriate string
    if guess_me < 7:
    print('too low')
    elif guess_me > 7:
    print('too high')
    else:
    print('just right')
    
    Output:
    just right
    
    As you can see, the string 'just right' is printed because guess_me is equal to 7. If guess_me was less than 7,
    the string 'too low' would have been printed, and if guess_me was greater than 7, the string 'too high' would ,
    have been printed.

    


##2. Assign the value 7 to the variable guess_me and the value 1 to the variable start. 
    #Write a while loop that compares start with guess_me. Print too low if start is less than guess me. 
    #If start equals guess_me,print 'found it!' and exit the loop. If start is greater than guess_me, print 'oops' ,
    #and exit the loop. Increment start at the end of the loop.

Ans-.Here's how you can assign the value 7 to guess_me and the value 1 to start,and then write the while loop to compare,
    start with guess_me and print the appropriate string based on the comparison:
    
    # Assign the value 7 to guess_me and the value 1 to start
    guess_me = 7
    start = 1

    # Write the while loop to compare start with guess_me and print the appropriate string
    while start <= guess_me:
    if start < guess_me:
        print('too low')
    elif start == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    start += 1

    Output:
        
    too low
    too low
    too low
    too low
    too low
    too low
    found it!
    
    As you can see, the while loop is executed until start is incremented to 7, which is equal to guess_me. At this point,
    the string 'found it!' is printed and the loop is exited. If start was less than guess_me,  the string 'too low' would ,
    be printed, and if start was greater than guess_me, the string 'oops' would be printed.
  
        
##3. Print the following values of the list [3, 2, 1, 0] using a for loop.

Ans-.# Define the list
    my_list = [3, 2, 1, 0]

    # Use a for loop to print each value of the list
    for value in my_list:
    print(value)
    
    Output:

    3
    2
    1
    0

    As you can see, the for loop iterates over each value in the list and prints it on a separate line.

##4. Use a list comprehension to make a list of the even numbers in range(10)

Ans-.Here's how you can use a list comprehension to create a list of even numbers in the range of 10:
    # Create a list of even numbers in the range of 10 using a list comprehension
    even_numbers = [num for num in range(10) if num % 2 == 0]

    # Print the resulting list
    print(even_numbers)
    
    Output:
        
    [0, 2, 4, 6, 8]

    
    In the list comprehension, we iterate over the range of numbers from 0 to 9, and only add the number to the list,
    if it is divisible by 2 (i.e., if it is even). The resulting list contains all the even numbers in the range of 10.



##5. Use a dictionary comprehension to create the dictionary squares. Use range(10) to return the keys, 
     #and use the square of each key as its value.

Ans-.Here is how you can use a dictionary comprehension to create a dictionary called squares:
    # Create a dictionary comprehension to create the dictionary 'squares'
    squares = {num: num**2 for num in range(10)}

    # Print the resulting dictionary
    print(squares)

    Output:
        
    {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64, 9: 81}

    In the dictionary comprehension, we iterate over the range of numbers from 0 to 9, and use each number as both the,
    key and the value of the dictionary. We calculate the square of each number by raising it to the power of 2 using,
    the ** operator. The resulting dictionary contains the squares of all the numbers from 0 to 9, with each number as its key.

##6. Construct the set odd from the odd numbers in the range using a set comprehension (10).

Ans-Here is how you can use a set comprehension to create a set called odd with the odd numbers in the range from 0 to 9:
    # Create a set comprehension to create the set 'odd'
    odd = {num for num in range(10) if num % 2 != 0}
    
    # Print the resulting set
    print(odd)
    
    Output:
     {1, 3, 5, 7, 9}
    
    In the set comprehension, we iterate over the range of numbers from 0 to 9, and use a conditional statement to,
    include only the odd numbers in the resulting set. We check whether each number is odd by using the modulus operator,
    % to determine if it has a remainder when divided by 2. If the remainderis not equal to 0, we include the number in the set. 
    The resulting set contains all the odd numbers from 0 to 9.



##7. Use a generator comprehension to return the string 'Got ' and a number for the numbers in range(10),
    #Iterate through this by using a for loop.

Ans-.Here is an example of how to use a generator comprehension to return a string with a number for each number in the range(10):
     # Use a generator comprehension to create a generator object
    gen = ('Got ' + str(num) for num in range(10))

    # Iterate through the generator object using a for loop
    for item in gen:
    print(item)

    Output:
    got 0
    Got 1
    Got 2
    Got 3
    Got 4
    Got 5
    Got 6
    Got 7
    Got 8
    Got 9
   
    In this example, we create a generator comprehension that concatenates the string 'Got ' with each number in the range from 
    0 to 9. We use the str() function to convert each number to a string before concatenating it with the string 'Got '.
    The generator comprehension returns a generator object that can be iterated through using a for loop. 
    We then iterate through the generator object and print each item, which consists of the string 'Got ' followed by a number.   
        
        
##8. Define a function called good that returns the list ['Harry', 'Ron', 'Hermione'].

Ans-.def good():
    return ['Harry', 'Ron', 'Hermione']

    This function simply returns a list of three strings, 'Harry', 'Ron', and 'Hermione', when called.


##9. Define a generator function called get_odds that returns the odd numbers from range(10). Use a for loop to find,
    #and print the third value returned.

Ans-.def get_odds():
    for number in range(1, 10, 2):
        yield number
    
    This generator function uses a for loop to iterate through the odd numbers in the range range(1, 10, 2) and yield each 
    number as it's generated.

    To find and print the third value returned by this generator function, you can use a for loop along with the next() function
    to iterate through the ,first three generated odd numbers:
        
    # Define the generator function
    def get_odds():
    for number in range(1, 10, 2):
        yield number

    # Iterate through the first three generated odd numbers
    count = 1
    for odd in get_odds():
    if count == 3:
        print("The third odd number is:", odd)
        break
    count += 1
    
    This code will output: The third odd number is: 5


##10. Define an exception called OopsException. Raise this exception to see what happens. 
    #Then write the code to catch this exception and print 'Caught an oops'.

Ans-.Here is the code to define and raise the OopsException and then catch it:
    # Define the exception
    class OopsException(Exception):
    pass

    # Raise the exception
    raise OopsException

    # Catch the exception
    try:
    raise OopsException
    except OopsException:
    print('Caught an oops')
    
    When running this code, the output will be:
        
    Traceback (most recent call last):
  File "<input>", line 3, in <module>
__main__.OopsException
    Caught an oops


##11. Use zip() to make a dictionary called movies that pairs these lists: titles = ['Creature of Habit', 'Crewel Fate'] ,
    and plots = ['A nun turns into a monster', 'A haunted yarn shop'].

Ans-.Here's how to use zip() to create the movies dictionary:

    titles = ['Creature of Habit', 'Crewel Fate']
    plots = ['A nun turns into a monster', 'A haunted yarn shop']

    movies = dict(zip(titles, plots))

    print(movies)
    
    Output:
        
    {'Creature of Habit': 'A nun turns into a monster', 'Crewel Fate': 'A haunted yarn shop'}

```
