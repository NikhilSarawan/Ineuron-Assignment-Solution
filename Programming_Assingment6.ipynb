```python
1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

Ans
    Function to generate Fibonacci sequence using recursion
    def fibonacci(n):
        if n <= 0:
            return "Invalid input"
        elif n == 1:
            return [0]
        elif n == 2:
            return [0, 1]
        else:
            fib_seq = fibonacci(n - 1)
            fib_seq.append(fib_seq[-1] + fib_seq[-2])
            return fib_seq

    Main function to take input from the user and display Fibonacci sequence
    if __name__ == "__main__":
        try:
            n = int(input("Enter the number of terms you want in Fibonacci sequence: "))
            fibonacci_sequence = fibonacci(n)
            print("Fibonacci sequence:")
            print(fibonacci_sequence)
        except ValueError:
            print("Invalid input. Please enter a valid number.")


2. Write a Python Program to Find Factorial of Number Using Recursion?

Ans
    Function to calculate factorial using recursion
    def factorial(n):
        if n == 0:
            return 1
        else:
            return n * factorial(n - 1)

    Main function to take input from the user and display factorial
    if __name__ == "__main__":
        try:
            num = int(input("Enter a non-negative integer: "))
            if num < 0:
                print("Factorial is not defined for negative numbers.")
            else:
                result = factorial(num)
                print(f"The factorial of {num} is: {result}")
        except ValueError:
            print("Invalid input. Please enter a valid non-negative integer.")


3. Write a Python Program to calculate your Body Mass Index?

Ans
    Function to calculate Body Mass Index (BMI)
    def calculate_bmi(weight, height):
        bmi = weight / (height ** 2)
        return bmi

    Main function to take input from the user and calculate BMI
    if __name__ == "__main__":
        try:
            weight = float(input("Enter your weight in kilograms: "))
            height = float(input("Enter your height in meters: "))
        
            if weight <= 0 or height <= 0:
                print("Invalid input. Please enter valid weight and height.")
            else:
                bmi = calculate_bmi(weight, height)
                print(f"Your Body Mass Index (BMI) is: {bmi:.2f}")
        except ValueError:
            print("Invalid input. Please enter valid numerical values for weight and height.")


4. Write a Python Program to calculate the natural logarithm of any number?

Ans
    import math

    Main function to take input from the user and calculate natural logarithm
    if __name__ == "__main__":
        try:
            num = float(input("Enter a positive number: "))
        
            if num <= 0:
                print("Invalid input. Please enter a positive number.")
            else:
                natural_log = math.log(num)
                print(f"The natural logarithm of {num} is: {natural_log:.2f}")
        except ValueError:
            print("Invalid input. Please enter a valid number.")


5. Write a Python Program for cube sum of first n natural numbers?

Ans
    Function to calculate cube sum of first n natural numbers
    def cube_sum(n):
        sum = 0
        for i in range(1, n + 1):
            sum += i ** 3
        return sum

    Main function to take input from the user and calculate cube sum
    if __name__ == "__main__":
        try:
            n = int(input("Enter a positive integer: "))
        
            if n <= 0:
                print("Invalid input. Please enter a positive integer.")
            else:
                result = cube_sum(n)
                print(f"The cube sum of first {n} natural numbers is: {result}")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")
```
