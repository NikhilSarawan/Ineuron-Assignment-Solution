```python
Question 1:
Write a program that calculates and prints the value according to the given formula:
Q = Square root of [(2 * C * D)/H]
Following are the fixed values of C and H:
C is 50. H is 30.
D is the variable whose values should be input to your program in a comma-separated
sequence.

Ans-.

import math

Fixed values of C and H
C = 50
H = 30

Function to calculate Q based on the given formula
def calculate_Q(D_values):
    Q_values = []
    for D in D_values:
        Q = math.sqrt((2 * C * D) / H)
        Q_values.append(Q)
    return Q_values

Input for variable D in a comma-separated sequence
D_values = input("Enter values of D separated by commas: ").split(',')

Convert input values from strings to integers
D_values = [int(d) for d in D_values]

Calculate and print the values of Q
Q_values = calculate_Q(D_values)
print("Values of Q corresponding to the given D values:")
print(Q_values)


Question 2:
Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
element value in the i-th row and j-th column of the array should be i*j.
Note: i=0,1.., X-1; j=0,1,Â¡Y-1.
    
Ans-

Function to generate 2-dimensional array based on input X and Y
def generate_2d_array(X, Y):
    Using list comprehension to create the 2-dimensional array
    Element at i-th row and j-th column is i * j
    return [[i * j for j in range(Y)] for i in range(X)]

Taking input for X and Y
X = int(input("Enter the number of rows (X): "))
Y = int(input("Enter the number of columns (Y): "))

Generating the 2-dimensional array
result_array = generate_2d_array(X, Y)

Printing the generated 2-dimensional array
print("Generated 2-dimensional array:")
for row in result_array:
    print(row)

Question 3:
Write a program that accepts a comma separated sequence of words as input and prints the
words in a comma-separated sequence after sorting them alphabetically.
 
Ans-
    
Taking input for comma-separated sequence of words
input_words = input("Enter a comma-separated sequence of words: ")

Splitting the input into a list of words
words_list = input_words.split(',')

Sorting the list of words alphabetically
sorted_words = sorted(words_list)

Joining the sorted words into a comma-separated sequence
sorted_words_sequence = ','.join(sorted_words)

Printing the sorted words in a comma-separated sequence
print("Words in sorted comma-separated sequence:", sorted_words_sequence)

Question 4:
Write a program that accepts a sequence of whitespace separated words as input and prints
the words after removing all duplicate words and sorting them alphanumerically.

Ans-

Taking input for whitespace-separated words
input_words = input("Enter a sequence of whitespace-separated words: ")

Splitting the input into a list of words
words_list = input_words.split()

Removing duplicate words using set() and sorting them alphanumerically
unique_sorted_words = sorted(set(words_list))

Joining the unique sorted words into a whitespace-separated sequence
result_sequence = ' '.join(unique_sorted_words)

Printing the unique sorted words in a whitespace-separated sequence
print("Words after removing duplicates and sorting them alphanumerically:")
print(result_sequence)


Question 5:
Write a program that accepts a sentence and calculate the number of letters and digits.
Suppose the following input is supplied to the program:
    
Ans-

Taking input for the sentence
input_sentence = input("Enter a sentence: ")

Initializing counters for letters and digits
letter_count = 0
digit_count = 0

Iterating through each character in the input sentence
for char in input_sentence:
    Checking if the character is a letter
    if char.isalpha():
        letter_count += 1
    Checking if the character is a digit
    elif char.isdigit():
        digit_count += 1

Printing the number of letters and digits in the input sentence
print("Number of letters: ", letter_count)
print("Number of digits: ", digit_count)


Question 6:
A website requires the users to input username and password to register. Write a program to
check the validity of password input by users.
Following are the criteria for checking the password:
1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
1. At least 1 letter between [A-Z]
3. At least 1 character from [$#@]
4. Minimum length of transaction password: 6
5. Maximum length of transaction password: 12
Your program should accept a sequence of comma separated passwords and will check them
according to the above criteria. Passwords that match the criteria are to be printed, each
separated by a comma.
                              
Ans-
                              
import re

Function to check the validity of passwords
def is_valid_password(password):
    Using regular expressions to check the criteria
    if re.match(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[$#@]).{6,12}$", password):
        return True
    else:
        return False

Taking input for comma-separated passwords
passwords = input("Enter comma-separated passwords: ").split(',')

Checking and printing valid passwords separated by commas
valid_passwords = [password for password in passwords if is_valid_password(password)]
print("Valid passwords:", ",".join(valid_passwords))
                              
```
