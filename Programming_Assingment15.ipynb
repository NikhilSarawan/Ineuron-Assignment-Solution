```python
Question 1:
Please write a program using generator to print the numbers which can be divisible by 5 and
7 between 0 and n in comma separated form while n is input by console.

Ans-

def divisible_by_5_and_7_generator(n):
    for num in range(n+1):
        if num % 5 == 0 and num % 7 == 0:
            yield num

Get input from the user
n = int(input("Enter a number: "))

Generate numbers divisible by 5 and 7 using the generator function
result_generator = divisible_by_5_and_7_generator(n)

Print the numbers in comma-separated form
print("Numbers divisible by 5 and 7 between 0 and", n, "are:", end=" ")
print(*result_generator, sep=", ")

Question 2:
Please write a program using generator to print the even numbers between 0 and n in comma
separated form while n is input by console.

Ans-

def even_numbers_generator(n):
    for num in range(n+1):
        if num % 2 == 0:
            yield num

Get input from the user
n = int(input("Enter a number: "))

Generate even numbers using the generator function
result_generator = even_numbers_generator(n)

Print the even numbers in comma-separated form
print("Even numbers between 0 and", n, "are:", end=" ")
print(*result_generator, sep=", ")

Question 3:
The Fibonacci Sequence is computed based on the following formula:
f(n)=0 if n=0
f(n)=1 if n=1
f(n)=f(n-1)+f(n-2) if n&gt;1
Please write a program using list comprehension to print the Fibonacci Sequence in comma
separated form with a given n input by console.

Ans-

Get input from the user
n = int(input("Enter a number: "))

Fibonacci sequence using list comprehension
fibonacci_sequence = [0 if i == 0 else 1 if i == 1 else fibonacci_sequence[i-1] + fibonacci_sequence[i-2] for i in range(n+1)]

Print the Fibonacci sequence in comma-separated form
print("Fibonacci sequence up to", n, "is:", end=" ")
print(*fibonacci_sequence, sep=", ")


Question 4:
Assuming that we have some email addresses in the &quot;username@companyname.com&quot; format,
please write program to print the user name of a given email address. Both user names and
company names are composed of letters only.


Ans-

Get email address input from the user
email_address = input("Enter an email address: ")

Split the email address at "@" to separate username and domain
username, domain = email_address.split("@")

Print the extracted username
print("Username:", username)

Question 5:
Define a class named Shape and its subclass Square. The Square class has an init function
which takes a length as argument. Both classes have a area function which can print the area
of the shape where Shape&#39;s area is 0 by default.

Ans-

class Shape:
    def __init__(self):
        pass

    def area(self):
        return 0

class Square(Shape):
    def __init__(self, length):
        super().__init__()
        self.length = length

    def area(self):
        return self.length ** 2

Example usage
length = float(input("Enter the length of the square: "))
square = Square(length)

print("Area of the square:", square.area())




```
