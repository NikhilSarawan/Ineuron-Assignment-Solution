```python
##1. To what does a relative path refer?

Ans-.A relative path refers to the path to a file or directory relative to the current working directory of the program.
    In other words, it specifies the location of a file or directory in relation to the directory that the program is currently,
    running in.

    For example, if a program is currently running in the directory "C:\Users\John\Documents", and there is a file named,
    "example.txt" located in a subdirectory called "data", then the relative path to that file would be "data/example.txt".
    The program can use this relative path to access the file, regardless of its absolute path on the system.

    Relative paths can be useful in situations where the absolute path to a file or directory may change,
    but the relationship between the file/directory and the program's working directory remains the same.

##2. What does an absolute path start with your operating system?

Ans-.An absolute path starts with the root directory of the operating system, followed by a series of subdirectories that lead to the desired file or directory. The root directory is the top-level directory of the file system, 
    and its name varies depending on the operating system.

    For example, on Windows, an absolute path might start with "C:" to indicate the root directory of the C: drive, 
    while on Linux or macOS, it might start with "/" to indicate the root directory of the entire file system.

    An example of an absolute path on Windows might be "C:\Users\John\Documents\example.txt", which specifies the full,
    path to a file named "example.txt" located in the "Documents" directory of the "John" ,
    user's home directory on theC: drive.

##3. What do the functions os.getcwd() and os.chdir() do?

Ans-.The os.getcwd() and os.chdir() functions are used to get and change the current working directory of a Python program,
    respectively.

    os.getcwd() returns a string representing the current working directory, which is the directory that the program is currently running in.
    Here's an example of how to use it:
    import os

    # get the current working directory
    current_dir = os.getcwd()

    # print the current working directory
    print(current_dir)
    
    This code will print out the current working directory of the program.

    os.chdir() is used to change the current working directory to a new location. Here's an example:
    
    import os

    # change the current working directory
    os.chdir('C:/Users/John/Documents')

    # get the new current working directory
    new_dir = os.getcwd()

    # print the new current working directory
    print(new_dir)
    
    This code changes the current working directory to the "Documents" directory of the,
    "John" user's home directory on the C: drive (on Windows), and then prints out the new current working directory.

    Note that os.chdir() changes the working directory for the entire program, so subsequent file operations will be relative,
    to the new working directory.

    
    


##4. What are the . and .. folders?

Ans-.In most file systems, the . and .. folders are special directories used to refer to the current directory and the ,
    directory, respectively.
    The . folder represents the current directory. For example, if a Python script is located in the directory,
    C:\Users\John\Documents\Scripts, and it needs to access a file located in the same directory, the file path,
    can be specified as .\filename.txt, where . represents the current directory.

    The .. folder represents the parent directory. For example, if the same Python script needs to access a file located,
    in the parent directory, the file path can be specified as ..\filename.txt, where .. represents the parent directory.

    Here's an example:
    import os

    
    import os

    # get the current working directory
    current_dir = os.getcwd()

    # create a new file in the current directory
    with open('example.txt', 'w') as f:
    f.write('This is an example file.')

    # navigate to the parent directory
    os.chdir('..')

    # list the contents of the parent directory
    print(os.listdir('.'))
    This code creates a new file called "example.txt" in the current working directory, and then navigates to the parent,
    directory using the .. directory. Finally,it lists the contents of the parent directory using os.listdir(), 
    which should include the "Documents" directory (assuming that the program was originally running in the "Documents",
    directory).

    Note that the . and .. directories are not actually physical directories on disk, but rather special names used to refer,
    to the current and parent directories, respectively.
    
    

##5. In C:\bacon\eggs\spam.txt, which part is the dir name, and which part is the base name?

Ans-.

##6. What are the three “mode” arguments that can be passed to the open() function?

Ans-.The open() function in Python is used to open a file and return a file object, which can be used to read from or write,
    to the file. The function takes several arguments, including a filename and a "mode" argument that specifies how the file,
    should be opened.

    There are three main mode arguments that can be passed to the open() function:

    'r' (read mode): This mode is used to open a file for reading. If the file does not exist, an error is raised.
    Here's an example:
    with open('example.txt', 'r') as f:
    data = f.read()
    print(data)
    
    This code opens a file called "example.txt" in read mode, reads the contents of the file using f.read(), and then prints,
    the contents to the console.

    'w' (write mode): This mode is used to open a file for writing. If the file does not exist, it will be created.
    If the file already exists, its contents will be overwritten. Here's an example
    
    
    with open('example.txt', 'w') as f:
    f.write('This is an example file.')
    
    This code opens a file called "example.txt" in write mode, writes the string "This is an example file." to the,
    file using f.write(), and then closes the file.

    'a' (append mode): This mode is used to open a file for writing, but instead of overwriting the file's contents,
    new data is appended to the end of the file. If the file does not exist, it will be created. Here's an example:
    
    with open('example.txt', 'a') as f:
    f.write('\nThis is some additional text.')
    
    This code opens a file called "example.txt" in append mode, writes the string "\nThis is some additional text." to the end of the file using f.write(), and then closes the file.

    Note that there are also several other mode arguments that can be used with the open() function, such as 'rb' for ,
    reading a binary file, or 'x' for creating a new file .
    (but raising an error if the file already exists).



##7. What happens if an existing file is opened in write mode?

Ans-.If an existing file is opened in write mode using the open() function in Python, its contents will be completely overwritten by any data that is written to it. Any existing data in the file will be lost.

    Here's an example:
    
    # open a file in write mode
    with open('example.txt', 'w') as f:
    # write some text to the file
    f.write('This is some text.')

    # open the same file in write mode again
    with open('example.txt', 'w') as f:
    # write some new text to the file
    f.write('This is some new text.')

    # read the contents of the file
    with open('example.txt', 'r') as f:
    data = f.read()
    print(data)
    
    In this code, a file called "example.txt" is opened in write mode using open() and some text is written to the file using,
    f.write(). The file is then closed. Next, the same file is opened again in write mode, and some new text is written to the,
    file. Finally, the contents of the file are read using open() and printed to the console using print().

    When the code is run, the output will be:
    This is some new text.
    
    This is because the second call to open() in write mode overwrote the contents of the file, replacing the original,
    text with the new text.

    If you want to add data to an existing file without overwriting its contents, you can use the "append" mode by,
    specifying 'a' as the mode argument to open() instead of 'w'.


##8. How do you tell the difference between read() and readlines()?

Ans-.In Python, both the read() and readlines() methods are used to read data from a file. However, there are some differences,
    in how they work.

    The read() method reads the entire contents of a file and returns them as a string. It takes an optional argument that,
    specifies the number of characters to read. If no argument is provided, the entire file is read. Here's an example:
    
    with open('example.txt', 'r') as f:
    data = f.read()
    print(data)
    
    In this code, the read() method is used to read the entire contents of a file called "example.txt" into the variable data. 
    The contents of the file are then printed to the console using print().

    On the other hand, the readlines() method reads the entire contents of a file and returns them as a list of strings, 
    where each string represents a line in the file. Here's an example
    
    with open('example.txt', 'r') as f:
    data = f.read()
    print(data)
    
    In this code, the read() method is used to read the entire contents of a file called "example.txt" into the variable data.
    The contents of the file are then printed to the console using print().

    On the other hand, the readlines() method reads the entire contents of a file and returns them as a list of strings,
    where each string represents a line in the file. Here's an example:
    
    with open('example.txt', 'r') as f:
    data = f.readlines()
    print(data)
    
    
    with open('example.txt', 'r') as f:
    data = f.readlines()
    print(data)
    
    In this code, the readlines() method is used to read the entire contents of a file called "example.txt" into,
    the variable data. The contents of the file are then printed to the console using print().

    The main difference between read() and readlines() is that read() returns the entire contents of the ,
    file as a single string, whereas readlines() returns the contents of the file as a list of strings, 
    where each string represents a line in the file.

    For example, if the contents of "example.txt" were
    
    Line 1
    Line 2
    Line 3
    
    hen read() would return the string:


    Line 1
    Line 2
    Line 3
    
    And readlines() would return the list:

    ['Line 1\n', 'Line 2\n', 'Line 3\n']
 
    Note that each string in the list ends with a newline character (\n), which represents the end of a line in the file.



##9. What data structure does a shelf value resemble?

Ans-.In Python, a shelf is a persistent, dictionary-like object that is stored as a file on disk. It is implemented using the,
    shelve module, which provides a simple interface for storing and retrieving data using keys.

    A shelf value resembles a dictionary, as it stores data in a key-value format. However, unlike a regular dictionary, 
    a shelf value is persistent, which means that it can be saved to disk and retrieved later.

    Here's an example of how to create and use a shelf value in Python
    
    import shelve

    # create a new shelf file
    with shelve.open('mydata') as shelf:
    # store some data in the shelf
    shelf['name'] = 'Alice'
    shelf['age'] = 30
    shelf['city'] = 'New York'

    # open the same shelf file and read the data
    with shelve.open('mydata') as shelf:
    print(shelf['name'])
    print(shelf['age'])
    print(shelf['city'])
    
    In this code, a new shelf file is created using the shelve.open() method. 
    Some data is stored in the shelf using dictionary-like syntax, where the keys are strings and the values can be any,
    Python object. The shelf is then closed, which saves the data to disk.

    Later, the same shelf file is opened using shelve.open() again, and the data is read from the shelf using the keys.
    The output of the program will be:
 
    Alice
    30
    New 
    
    As you can see, the shelf value behaves like a dictionary, allowing you to store and retrieve data using keys.
    However, because it is persistent, you can save the data to disk and retrieve it later, making it a useful tool,
    for storing data between program executions.
```
