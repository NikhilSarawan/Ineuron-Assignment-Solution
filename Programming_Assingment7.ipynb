```python
1. Write a Python Program to find sum of array?

Ans
    Function to calculate the sum of elements in an array
    def find_sum(arr):
        sum = 0
        for num in arr:
            sum += num
        return sum

    Input: Taking array elements as input from the user
    Example input: 1 2 3 4 5 (space-separated numbers)
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))

    Calculating the sum using the function
    result = find_sum(arr)

    Output: Displaying the sum of the array elements
    print("Sum of the array elements is:", result)


2. Write a Python Program to find largest element in an array?

Ans
    Function to find the largest element in an array
    def find_largest_element(arr):
        Assuming the first element as the largest
        largest = arr[0]
        Iterating through the array to find the largest element
        for num in arr:
            if num > largest:
                largest = num
        return largest

    Input: Taking array elements as input from the user
    Example input: 1 2 3 4 5 (space-separated numbers)
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))

    Calculating the largest element using the function
    largest_element = find_largest_element(arr)

    Output: Displaying the largest element in the array
    print("The largest element in the array is:", largest_element)


3. Write a Python Program for array rotation?

Ans
    Function to rotate the array to the right by d positions
    def rotate_array(arr, d):
        n = len(arr)
    Using slicing to rotate the array
        rotated_array = arr[n - d:] + arr[:n - d]
        return rotated_array

    Input: Taking array elements as input from the user
    Example input: 1 2 3 4 5 (space-separated numbers)
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))
    Input: Taking the number of positions to rotate as input
    d = int(input("Enter the number of positions to rotate the array to the right: "))

    Rotating the array using the function
    rotated_array = rotate_array(arr, d)

    Output: Displaying the rotated array
    print("Array after rotating", d, "positions to the right:", rotated_array)


4. Write a Python Program to Split the array and add the first part to the end?

Ans
    Function to split the array and add the first part to the end
    def split_and_add(arr, k):
    Using slicing to split the array and add the first part to the end
        rotated_array = arr[k:] + arr[:k]
        return rotated_array

    Input: Taking array elements as input from the user
    Example input: 1 2 3 4 5 (space-separated numbers)
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))
    Input: Taking the index to split the array as input
    k = int(input("Enter the index to split the array: "))

    Splitting the array and adding the first part to the end using the function
    result_array = split_and_add(arr, k)

    Output: Displaying the modified array
    print("Array after splitting at index", k, "and adding the first part to the end:", result_array)


5. Write a Python Program to check if given array is Monotonic?

Ans
    Function to check if the array is monotonic
    def is_monotonic(arr):
    Flag variables to check for monotonicity
        increasing = decreasing = True
    
    Iterate through the array to check for monotonicity
        for i in range(1, len(arr)):
            if arr[i] > arr[i - 1]:
                decreasing = False
            elif arr[i] < arr[i - 1]:
                increasing = False
    
    If either increasing or decreasing is True, the array is monotonic
    return increasing or decreasing

    Input: Taking array elements as input from the user
    Example input: 1 2 3 4 5 (space-separated numbers)
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))

    Checking if the array is monotonic using the function
    monotonic = is_monotonic(arr)

    Output: Displaying the result
    if monotonic:
        print("The given array is monotonic.")
    else:
        print("The given array is not monotonic.")
```
