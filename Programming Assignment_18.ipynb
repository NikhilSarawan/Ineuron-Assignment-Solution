```python
Question 1
Create a function that takes a list of non-negative integers and strings and return a new list
without the strings.

Ans-

def filter_list(lst):
    return [x for x in lst if isinstance(x, int) and x >= 0]

Examples
print(filter_list([1, 2, "a", "b"]))  # ➞ [1, 2]
print(filter_list([1, "a", "b", 0, 15]))  # ➞ [1, 0, 15]
print(filter_list([1, 2, "aasf", "1", "123", 123]))  # ➞ [1, 2, 123]



Question 2
The &quot;Reverser&quot; takes a string as input and returns that string in reverse order, with the
opposite case.

Ans-

def reverse(input_string):
Reverse the string and change the case using slicing and str.swapcase()
    reversed_string = input_string[::-1].swapcase()
    return reversed_string

Examples
print(reverse("Hello World"))  # ➞ "DLROw OLLEh"
print(reverse("ReVeRsE"))      # ➞ "eSrEvEr"
print(reverse("Radar"))         # ➞ "RADAr"


Question 3
You can assign variables from lists like this:

Ans-

Given list
writeyourcodehere = [1, 2, 3, 4, 5, 6]

Destructuring assignment
first, *middle, last = writeyourcodehere

Printing the variables
print("first ➞", first)  # Output: first ➞ 1
print("middle ➞", middle)  # Output: middle ➞ [2, 3, 4, 5]
print("last ➞", last)  # Output: last ➞ 6


Question 4
Write a function that calculates the factorial of a number recursively.

Ans-

def factorial(n):
    Base case: factorial of 0 or 1 is 1
    if n == 0 or n == 1:
        return 1
    Recursive case: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

Examples
print(factorial(5))  # ➞ 120
print(factorial(3))  # ➞ 6
print(factorial(1))  # ➞ 1
print(factorial(0))  # ➞ 1


Question 5
Write a function that moves all elements of one type to the end of the list.

Ans-

def move_to_end(lst, element):
    # Partition the list into elements equal to and not equal to the given element
    equal_elements = [x for x in lst if x == element]
    non_equal_elements = [x for x in lst if x != element]
    # Concatenate the non-equal elements and equal elements to form the final list
    return non_equal_elements + equal_elements

Examples
print(move_to_end([1, 3, 2, 4, 4, 1], 1))  # ➞ [3, 2, 4, 4, 1, 1]
print(move_to_end([7, 8, 9, 1, 2, 3, 4], 9))  # ➞ [7, 8, 1, 2, 3, 4, 9]
print(move_to_end(["a", "a", "a", "b"], "a"))  # ➞ ["b", "a", "a", "a"]
```
